```python
import os

def print_project_tree(dir_path, prefix=""):
    """
    Recursively maps out the project structure, 
    ignoring massive virtual environments and hidden caches.
    """
    # 1. The Filter: Add any other noisy folders you want to hide in here
    ignore_items = {
        '.git', '__pycache__', '.venv', 'venv', 'env', 
        'node_modules', '.ipynb_checkpoints', 'mb3_model_cache'
    }
    
    try:
        # Get all files and folders, and sort them
        items = sorted(os.listdir(dir_path))
    except PermissionError:
        return

    # Filter out the noise
    items = [item for item in items if item not in ignore_items]
    
    # 2. The Formatter: Draws the actual tree lines
    for index, item in enumerate(items):
        path = os.path.join(dir_path, item)
        is_last = (index == len(items) - 1)
        connector = "└── " if is_last else "├── "
        
        print(f"{prefix}{connector}{item}")
        
        # If it's a folder, dive into it recursively
        if os.path.isdir(path):
            extension = "    " if is_last else "│   "
            print_project_tree(path, prefix=prefix + extension)

# 3. Execution
print(f"📁 {os.path.basename(os.getcwd())}/")
print_project_tree(".")
```

```python
import os
from pathlib import Path

def setup_production_structure():
    # 1. Define the directory tree
    folders = [
        "data/markdown_vault",
        "data/scanned_textbooks",
        "data/slides",
        "experiments",
        "backend/alembic/versions",
        "backend/src/api",
        "backend/src/core",
        "backend/src/db",
        "backend/src/models",
        "backend/src/schemas",
        "backend/src/services",
    ]

    # 2. Define the essential files (empty as requested)
    files = [
        # Root files
        ".env",
        ".gitignore",
        "README.md",
        # Backend Root
        "backend/alembic.ini",
        "backend/Dockerfile",
        "backend/docker-compose.yml",
        # Source files
        "backend/src/main.py",
        "backend/src/prompts.py",
        "backend/src/core/config.py",
        "backend/src/core/security.py",
        "backend/src/db/session.py",
        "backend/src/models/user.py",
        "backend/src/models/history.py",
        "backend/src/schemas/user.py",
        "backend/src/schemas/chat.py",
        "backend/src/services/rag_engine.py",
        "backend/src/services/tools.py",
        "backend/src/api/auth.py",
        "backend/src/api/sessions.py",
        "backend/src/api/agent.py",
    ]

    print("🚀 Starting MB_ASSISTANT production setup...")

    # 3. Create Folders and __init__.py files
    for folder in folders:
        path = Path(folder)
        path.mkdir(parents=True, exist_ok=True)
        print(f"📁 Folder ready: {folder}")
        
        # Add __init__.py to all src subfolders for package recognition
        if "src" in folder or "api" in folder:
            init_file = path / "__init__.py"
            init_file.touch(exist_ok=True)

    # 4. Create empty Files
    for file_path in files:
        path = Path(file_path)
        if not path.exists():
            path.touch()
            print(f"📄 Created file: {file_path}")
        else:
            print(f"🟡 Skipping (already exists): {file_path}")

    print("\n✅ Setup Complete! You now have a modular structure ready for production.")

if __name__ == "__main__":
    setup_production_structure()
```
